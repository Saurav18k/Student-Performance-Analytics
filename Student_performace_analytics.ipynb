Calculate Total marks , Grade and Percentage . 

```python
def calculate_grades_and_performance(df):
    # Total aur Percentage calculate
    
    df['Total Marks'] = df['Maths'] + df['Science'] + df['English']
    df['Percentage'] = ((df['Total Marks'] / 300) * 100).round(2)
    df['Percentage'] = df['Percentage'].astype(float).round(2)
    # Grading Logic
    def assign_grade(pct):
        if pct >= 85: return 'A'
        elif pct >= 70: return 'B'
        elif pct >= 50: return 'C'
        else: return 'D'
        
    df['Grade'] = (df['Percentage'].apply(assign_grade))
    
    # Attendance Status Logic (>= 90 Good Attendance, else Needs Improvement)
    df['Attendance Status'] = df['Attendance'].apply(lambda x: 'Good Attendance' if x >= 90 else 'Needs Improvement')
    return df
```
